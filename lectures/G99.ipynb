### Parameterizing eddy flux in the Southern Ocean

*Parametrization*: Representing complex small scale processes with a simplified approximation. 

Remember, 
\begin{eqnarray}
F_b^y&=&<\overline{v'b'}>+<v^{\dagger}b^{\dagger}>\\
F_b^z&=&<\overline{w'b'}>+<w^{\dagger}b^{\dagger}>
\end{eqnarray}
are the eddy fluxes of buoyancy in the zonal and meridional directions.

Gent and McWilliams (1990) suggested that we should parameterize the eddy buoyancy flux as
\begin{equation}
    F_b^y=-K_{ed}\overline{b_y} \, .
\end{equation}
Don't worry too much about why: we mostly use this because it's simple and it works ok. Using this parametrization, the eddy transport per unit width is
\begin{eqnarray}
    \psi_{eddy}&=&\frac{F_b^y}{\overline{b_z}}\\
   &=&-K_{ed}\frac{\overline{b_y}}{\overline{b_z}} \\
    &=&K_{ed}s_{\rho} \, , 
\end{eqnarray}
where $s_{\rho}$ is the isopycnal slope (remember, an isopycnal is a surface of constant density). Using this approximation, the steeper the isopycnals, the larger the eddy transport.

### Gnanadesikan's model (1999)

Gnanadesikan's model describes an imaginary world comprised of the Atlantic Ocean and the Southern Ocean, like this:


Ignoring the Pacific and Indian Oceans is a useful simplification: they are less important for the deep circulation, because no deep water is formed there. From the side, Gnanadesikan's model looks a bit like this:


In this figure, the thick line represents a mid-depth isopycnal that reaches the surface in the far south of the re-entrant channel. The vertical dashed line represents the northern edge of the re-entrant channel. Gnanadesikan's approach is to look at the residual transport into the region enclosed by the orange line: this region is bounded by the northern edge of the re-entrant channel and by a mid-depth isopycnal north of the channel. He suggests expressions for each of the transports into and out to this region, and then applies conservation of buoyancy.

First, the Ekman transport at the northern edge of the re-entrant channel is
\begin{equation}
    T_{Ek}=\oint -\frac{\tau}{\rho f_s}\, \mbox{d}x\approx-\frac{\overline{\tau} L_x}{\rho f_s}\, 
\end{equation}
where $\overline{\tau}$ is the zonally-averaged wind stress at the northern edge of the re-entrant channel, and $f_s$ is the Coriolis parameter at the northern edge of the re-entrant channel.

From section 1, the eddy transport
\begin{equation}
    T_{ed}\approx=K_{ed}s_{\rho}\, ,
\end{equation}
where $s_{\rho}$ is the slope of the chosen isopycnal. We can approximate this slope as $h/L_y$, giving
\begin{equation}
    T_{ed}=\oint K_{ed}\frac{h}{L_y}\, \mbox{d}x\approx K_{ed}\frac{h L_x}{L_y}\, .
\end{equation}

The vertical diapycnal transport obeys the vertical advection-diffusion equation, 
\begin{equation}
    \kappa_{\nu}\frac{\partial^2 b}{\partial z^2}=w\frac{\partial b}{\partial z} \, .
\end{equation}
This equation suggests that the vertical velocity $w$ can be approximated as $\kappa_{\nu}/h$, so the vertical diapycnal transport can be approximated using
\begin{equation}
    T_{\nu}=\int w\, \mbox{d}A=w A \approx \frac{\kappa_{\nu} A}{h} \, , 
\end{equation}
where $\kappa_{\nu}$ is the vertical diffusivity and $A$ is the area over which there is uniform vertical diapycnal upwelling. 

Finally, the deep water formation rate is assumed to be equal to the northward geostrophic transport in the North Atlantic:
\begin{equation}
    T_n=\int \int v_g \, \mbox{d}z\,  \mbox{d}x
\end{equation}
The geostrophic velocity obeys the equation, \begin{equation}
-fv_g=-\frac{1}{\rho}\frac{\partial p}{\partial x}=-g\frac{\delta\rho}{\rho}\frac{\partial h}{\partial x}=-g'\frac{\partial h}{\partial x} \, ,
\end{equation}
where $g^{\prime}=g\delta \rho/\rho$ is the reduced gravity between the two layers. Integrating with respect to $z$ and $x$, 
\begin{equation}
    T_n=\int \int v_g \, \mbox{d}z \,   \mbox{d}x=\frac{g'(h|_{x_e}^2-h|_{x_w}^2)}{2f_n} \, ,
\end{equation}
where $f_n$ is the Coriolis parameter just south of the location where deep water is formed, $x_e$ is the east side of the basin and $x_w$ is the west side of the basin. We assume $h|_{x_e}=h$ and $h|_{x_w}\approx 0$, so
\begin{equation}
    T_n=\frac{g'h^2}{2f_n} \, ,
\end{equation}

For buoyancy to be conserved, the total transport into the region enclosed by the orange line must be zero:
\begin{equation}
    T_{Ek}-T_{ed}+T_{\nu}=T_n \, , 
\end{equation}
so 
\begin{equation}
    -\frac{\overline{\tau} L_x}{\rho f_s}
    -K_{ed}\frac{h L_x}{L_y}+\frac{\kappa_{\nu} A}{h}=\frac{g^{\prime}h^2}{2f_n}
\end{equation}
By setting $g^{\prime}, \overline{\tau},f_s,f_n, \rho, \kappa_{\nu}, K_{ed}$, this model allows us to predict $h,T_n, T_{Ek},T_{ed},T_{\nu}$. ### Nikurashin and Vallis 2012

Nikurashin and Vallis's model also describes an imaginary world with only an Atlantic Ocean and a Southern Ocean. But instead of thinking about the depth of and transport across one isopycnal,  Nikurashin and Vallis's model predicts the zonally-averaged buoyancy and transport as a function of latitude and depth ($y$ and $z$). It also includes deep water formation in the Southern Ocean. The side view of their ocean looks like this:

It is divided into region 1: the re-entrant channel, region 2: the basin, and region 3: the region where NADW is formed. In each of these regions we try to find an equation for buoyancy and an equation for transport.

### Region 1: the re-entrant channel
In region 1, the surface Ekman transport per unit width is:
\begin{equation}
    \psi_{Ek}=-\frac{\tau}{\rho f} \, , 
\end{equation}
the eddy transport per unit width is
\begin{equation}
    \psi_{ed}=K_{ed}\frac{\partial_y b_1}{\partial_z b_1} \, , 
\end{equation}
and the total transport per unit width is given by $\psi_1=\psi_{Ek}+\psi_{ed}$. The buoyancy at the surface of region 1 is relaxed towards a set profile $b_s$:
\begin{equation}
    \kappa_{\nu}\left.\frac{\partial b_1}{\partial z}\right|_{z=0}=\lambda(b_s-b_1) \, .
\end{equation}
This is supposed to mimic the way the ocean temperature responds to atmospheric forcing. The buoyancy in region 1 is governed by the advection-diffusion equation in the $y-z$ plane, so
\begin{equation}
    \frac{\partial b_1}{\partial t}+v_1\frac{\partial b_1}{\partial y}+w_1\frac{\partial b_1}{\partial z}=\frac{\partial}{\partial y}\left(\kappa_h\frac{\partial b_1}{\partial y}\right)+\frac{\partial}{\partial z}\left(\kappa_{\nu}\frac{\partial b_1}{\partial z}\right) \, .
\end{equation}
Using
\begin{equation}
    v_1=-\frac{\partial \psi_1}{\partial z} \qquad \mbox{and}\qquad w_1=\frac{\partial \psi_1}{\partial y}\, , 
\end{equation}
setting $\kappa_h\approx 0$ and assuming steady state we find that
\begin{equation}
    \frac{\partial \psi_1}{\partial y}\frac{\partial b_1}{\partial z}-\frac{\partial \psi_1}{\partial z}\frac{\partial b_1}{\partial y}=\frac{\partial}{\partial z}\left(\kappa_{\nu}\frac{\partial b_1}{\partial z}\right)
\end{equation}

### Region 2: the basin
In the basin, isopycnals are assumed to be flat, so $\partial_y b_2=0$, and the advection-diffusion equation gives
\begin{equation}
    \frac{\partial \psi_2}{\partial y}\frac{\partial b_2}{\partial z}=\frac{\partial}{\partial z}\left(\kappa_{\nu}\frac{\partial b_2}{\partial z}\right) \, . 
\end{equation}
In the basin, upwelling is approximately uniform in the $y$ direction, so
\begin{equation}
  \frac{\partial \psi_2}{\partial y}=w_2=\frac{\left.\psi_3\right|_{y=L_y}-\left.\psi_1\right|_{y=0}}{L_y}  \, .
\end{equation}
### Region 3: the region where NADW is formed
Nikurashin and Vallis take a different approach to Gnanadesikan in the north of the basin. The isopycnals continue to be horizontal until they join on to surface boundary condition vertically. Nikurashin and Vallis assume there is a zonal geostrophic current in the region where NADW is formed, so
\begin{equation}
    u_3=-\frac{1}{f}\int_{-H}^z \partial_y b_3\, dz'+C \, , 
\end{equation}
where $-H$ is the bottom of the ocean. This implies eastward flow near the surface and westward flow at depth. In order for mass to be conserved, these two flows must integrate to zero in the vertical:
\begin{equation}
    \int_{-H}^0u_3\, dz'=0 \, ,
\end{equation}
implying that there is a vertical downward flow near the eastern boundary of the "Atlantic" basin. 
Hence, 
\begin{equation}
    \psi_3=\int_{L_y}^{L_y+l_n} w_3 \, d \mbox{y}=\int_{-H}^z\int_{L_y}^{L_y+l_n} u_3 \, dy \, dz
\end{equation}

### Putting it all together
The transport and buoyancy are matched at the boundaries of each region: $\psi_1(y=0)=\psi_2(y=0)$, $b_1(y=0)=b_2(y=0)$ and $\psi_2(y=L_y)=\psi_3(y=L_y)$. The full set of equations must be solved numerically. Some of the results are shown below - they are consistent with more complex models of the AMOC and with observations.
